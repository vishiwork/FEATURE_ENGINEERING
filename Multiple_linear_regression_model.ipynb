MULTIPLE LINEAR REGRESSION MODEL

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('D:/workspace/Multiple-Linear-Regression-master/50_Startups.csv')
```

```python
df.head()
```

#To check if any null values

```python
df.isnull()
```

```python
sns.heatmap(df.isnull(),yticklabels=False)
```

```python
#profit is dependent variable and the other columns are independednt variables,splitting into 2 variables
```

```python
x=df.iloc[:,:-1]
y=df.iloc[:,-1]
```

```python
#To convert the state feature into integer value using one hot encoding
```

```python
X= pd.get_dummies(data=x.State,drop_first=True)
```

```python
x.drop('State',axis=1,inplace=True)
```

```python
pd.concat([x,X],axis=1)
```

```python
#Splitting the data set into training and testing model
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.2,random_state=0)
```

```python
#Fitting multiple linear regression to the training set
```

```python
from sklearn.linear_model import LinearRegression
```

```python
regressor =LinearRegression()
```

```python
regressor.fit(x_train,y_train)
```

```python
#Predicting the result
```

```python
y_pred = regressor.predict(x_test)
```

```python
y_pred
```

```python
#to check the accuracy using r2 score  --->closer to 1 better accuracy
```

```python
from sklearn.metrics import r2_score
```

```python
score = r2_score(y_test,y_pred)
```
