```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
import numpy as np
```

```python
x = np.arange(1,10)
y = np.arange(11,20)
```

```python
a = np.arange(40,50)
b = np.arange(60,80)
```

```python
#Scatter plot
```

```python
plt.scatter(x,y,c='g')
plt.xlabel('X AXIS')
plt.ylabel('Y AXIS')
plt.title("Scatter Plot")
```

```python
y = x*x
```

```python
#plot
```

```python
plt.plot(x,y,'r--*',linewidth =2,markersize =12)
```

```python
#CREATING SUBPLOT
```

```python
plt.subplot(2,2,1)
plt.plot(x,y,'g')
plt.subplot(2,2,2)
plt.plot(x,y,'r^')
```

```python
np.pi
```

```python
x =np.arange(0,4*np.pi,0.1)
y =np.sin(x)
plt.title("SINE WAVE")
plt.plot(x,y)
```

```python
#sin and cosine wave forms
```

```python
x = np.arange(0,5*np.pi,0.1)
y_sin = np.sin(x)
y_cos = np.cos(x)

plt.subplot(2,1,1)
plt.plot(x,y_sin)
plt.title('sine')

plt.subplot(2,1,2)
plt.plot(x,y_cos)
plt.title('cosine')
```

```python
#BAR PLOT
```

```python
x = [2,8,10]
y = [12,16,18]

x2 = [3,9,11]
y2 = [5,8,15]

plt.bar(x,y,align = 'center')
plt.bar(x2,y2,color='g')
plt.title('BAR CHART')
```
