```python
#JOINPLOT


```

```python
sns.jointplot(x=df['tip'],y =df['total_bill'],kind = 'scatter')
```

```python
#PAIRPLOT
```

```python
sns.pairplot(df,hue='smoker')
```

```python
#distplot
```

```python
sns.distplot(df['tip'],kde=True)
```

```python
#CATEGORICAL PLOTS
```

```python
#  violin plots,count plots , box plots
```

```python
#1.COUNT PLOTS
```

```python
sns.countplot(x='sex',data =df,hue ='smoker',palette='rainbow')
```

```python
#BOX PLOT
```

```python
sns.boxplot('sex','total_bill',data=df)
```

```python
sns.boxplot('day','total_bill',data=df,hue='sex',palette='rainbow')
```

```python
#VIOLIN PLOT
```

```python
sns.violinplot(x='day',y='total_bill',data=df)
```
